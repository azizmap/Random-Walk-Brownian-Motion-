```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def random_walk(x_min,y_min,x_max,y_max):
    xi = []
    yi = []
    
    
    x = random.randint(x_min,x_max)
    y = random.randint(y_min,y_max)
    
    x_range = x_max - x_min
    y_range = y_max - y_min
    xi.append(x)
    yi.append(y)
    
    n = 100 #banyak iterasi
    
    for i in range(n):
        rand = random.uniform(0,1)
        if rand <= 0.25:
            x+=1
        elif (rand<= 0.5):
            y-=1
        elif (rand <=0.75):
            x-=1
        else:
            y+=1
        #pbc
        if (x > x_max):
            x -= x_range
        if (x < x_min):
            x += x_range
        if (y > y_max):
            y -= y_range
        if (y < y_min):
            y +=y_range
        #update position
        xi.append(x)
        yi.append(y)
        
    return xi,yi
```
